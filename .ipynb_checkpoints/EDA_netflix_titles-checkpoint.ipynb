```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.info()
```

```python
# sütunların unique değerleri
for column in df.columns:
    print(column)
    print(df[column].value_counts())
    print("---" * 20)
```

```python
# sütunların eksik veri incelemesi
for column in df.columns:
    print("Column: ",column)
    print("Toplam null değerler: ",df[column].isnull().sum())
    print("---" * 20)
```

```python
sns.heatmap(df.isnull(),
            yticklabels=False,
            cbar=False,
            cmap='magma');
plt.title("Eksik Veri Gözlemi", fontsize = 15)
plt.show()
```

```python
# eksik değerleri doldurmak için sütun type abject olduğu için mode kullanabiliriz ancak bu veriseti için mantıklı olmayabilir 
# mode ile doldursaydık aşağıdaki işlem yapılabilirdi:
'''
for column in df.columns:
    print("Column: ", column)
    mask = df[column].isnull()

    if mask.any():
        print("\tNull değerler var --> ", column)
        mode_value = df[column].mode().iloc[0]
        print("\t\tMode değeri ile dolduruluyor --> ", mode_value)
        df[column].fillna(mode_value, inplace=True)
'''
```

```python
sns.countplot(data = df, x = "type")
plt.title("İçeriklerin Türleri",fontsize = 18)
plt.show()
```

```python
df["type"].value_counts().plot(kind = "pie", autopct = "%1.2f")
plt.title("Film ve Dizilerin Dağılımı",fontsize = 18)
plt.show()
```

```python
# En Çok Filmi Olan İlk 10 Yönetmen
df[df["type"] == "Movie"]["director"].value_counts()[:10]
```

```python
director_film = df[df["type"] == "Movie"]["director"].value_counts()[:10]

plt.figure(figsize = (16,10))
sns.barplot(x = director_film.index, y = director_film.values)
plt.xticks(rotation = 45, fontsize = 14)
plt.title("En Çok Filmi Olan İlk 10 Yönetmen",fontsize = 18)
plt.show()
```

```python
# En çok içerik üreten 10 ülke
top10 = df["country"].value_counts()[:10]
top10
```

```python
plt.figure(figsize = (16,10))
sns.barplot(x = top10.index, y = top10.values)
plt.title("En Çok İçerik Üreten 10 Ülke", fontsize = 22)
plt.xticks(size = 15)
plt.show()
```

```python
# datetime dönüşümü ve ay, yıl sütun ataması
df["date_added"] = pd.to_datetime(df["date_added"])
df["year_added"] = pd.to_datetime(df["date_added"]).dt.year
df["month_added"] = pd.to_datetime(df["date_added"]).dt.month
df.head(3)
```

```python
df[df["type"] == "Movie"].groupby(["year_added", "month_added"]).size()
```

```python
movie_counts = df[df["type"] == "Movie"].groupby(["year_added", "month_added"]).size().reset_index(name = "count")
movie_counts
```

```python
# yıllara ve aylara göre film dağılımları
movie_counts.pivot_table(values = "count", index = "year_added", columns = "month_added", fill_value = 0)
```

```python
heatmap_film = movie_counts.pivot_table(values = "count", index = "year_added", columns = "month_added", fill_value = 0)
plt.figure(figsize = (16,10))
sns.heatmap(heatmap_film, annot = True, fmt = "d", cmap = "coolwarm")
plt.title("Yıllara ve Aylara Göre Film Dağılımı", fontsize = 20)
plt.show()
```

```python
# yıllara göre type durum
df.groupby(["release_year", "type"]).size().reset_index(name="count")
```

```python
release_counts = df.groupby(["release_year", "type"]).size().reset_index(name="count")

sns.lineplot(data=release_counts, x="release_year",
             y="count", hue="type")
plt.title("Her Yıl Çıkan Dizi & Film Sayısı", fontsize = 16)
plt.show()
```

```python
content_counts = df.groupby(["year_added", "type"]).size().reset_index(name = "count")

sns.barplot(x = "year_added", y = "count", data = content_counts, hue = "type")
plt.xticks(rotation = 45)
plt.title("Netflix'e Her Yıl Eklenen Dizi & Film Sayısı",fontsize = 14)
plt.show()
```

```python
# 2019 yılı Sci-Fi & Fantasy türü yapımlar 
df[(df["release_year"] == 2019) & (df["listed_in"].str.contains("Sci-Fi & Fantasy"))][:5]
```

```python
Sci_Fi_Fantasy = df[(df["release_year"] == 2019) & (df["listed_in"].str.contains("Sci-Fi & Fantasy"))]

plt.figure(figsize = (12,6))
Sci_Fi_Fantasy["type"].value_counts().plot(kind = "pie", autopct = "%1.2f")
plt.title("Sci-Fi & Fantasy Türünün Film ve Show Dağılımı(2019)",fontsize = 18)
plt.show()
```

```python
# korku içerikleri filmler
df[df["listed_in"].isin(["Horror Movies","Thrillers"])].head()
```

```python
# korku içerikli filmelerin yıllara göre dağılımı
Horror = df[df["listed_in"].isin(["Horror Movies","Thrillers"])]["release_year"].value_counts()
Horror
```

```python
sns.barplot(x = Horror.index, y = Horror.values)
plt.ylabel("Count")
plt.xticks(rotation = 90)
plt.title("Yıllara göre Korku Film Dağılımı",fontsize = 18)
plt.show()
```
