```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.info()
```

```python
# s√ºtunlarƒ±n unique deƒüerleri
for column in df.columns:
    print(column)
    print(df[column].value_counts())
    print("---" * 20)
```

```python
# s√ºtunlarƒ±n eksik veri incelemesi
for column in df.columns:
    print("Column: ",column)
    print("Toplam null deƒüerler: ",df[column].isnull().sum())
    print("---" * 20)
```

```python
sns.heatmap(df.isnull(),
            yticklabels=False,
            cbar=False,
            cmap='magma');
plt.title("Eksik Veri G√∂zlemi", fontsize = 15)
plt.show()
```

```python
# eksik deƒüerleri doldurmak i√ßin s√ºtun type abject olduƒüu i√ßin mode kullanabiliriz ancak bu veriseti i√ßin mantƒ±klƒ± olmayabilir 
# mode ile doldursaydƒ±k a≈üaƒüƒ±daki i≈ülem yapƒ±labilirdi:
'''
for column in df.columns:
    print("Column: ", column)
    mask = df[column].isnull()

    if mask.any():
        print("\tNull deƒüerler var --> ", column)
        mode_value = df[column].mode().iloc[0]
        print("\t\tMode deƒüeri ile dolduruluyor --> ", mode_value)
        df[column].fillna(mode_value, inplace=True)
'''
```

```python
sns.countplot(data = df, x = "type")
plt.title("ƒ∞√ßeriklerin T√ºrleri",fontsize = 18)
plt.show()
```

```python
df["type"].value_counts().plot(kind = "pie", autopct = "%1.2f")
plt.title("Film ve Dizilerin Daƒüƒ±lƒ±mƒ±",fontsize = 18)
plt.show()
```

```python
# En √áok Filmi Olan ƒ∞lk 10 Y√∂netmen
df[df["type"] == "Movie"]["director"].value_counts()[:10]
```

```python
director_film = df[df["type"] == "Movie"]["director"].value_counts()[:10]

plt.figure(figsize = (16,10))
sns.barplot(x = director_film.index, y = director_film.values)
plt.xticks(rotation = 45, fontsize = 14)
plt.title("En √áok Filmi Olan ƒ∞lk 10 Y√∂netmen",fontsize = 18)
plt.show()
```

```python
# En √ßok i√ßerik √ºreten 10 √ºlke
top10 = df["country"].value_counts()[:10]
top10
```

```python
plt.figure(figsize = (16,10))
sns.barplot(x = top10.index, y = top10.values)
plt.title("En √áok ƒ∞√ßerik √úreten 10 √úlke", fontsize = 22)
plt.xticks(size = 15)
plt.show()
```

```python
# datetime d√∂n√º≈ü√ºm√º ve ay, yƒ±l s√ºtun atamasƒ±
df["date_added"] = pd.to_datetime(df["date_added"])
df["year_added"] = pd.to_datetime(df["date_added"]).dt.year
df["month_added"] = pd.to_datetime(df["date_added"]).dt.month
df.head(3)
```

```python
df[df["type"] == "Movie"].groupby(["year_added", "month_added"]).size()
```

```python
movie_counts = df[df["type"] == "Movie"].groupby(["year_added", "month_added"]).size().reset_index(name = "count")
movie_counts
```

```python
# yƒ±llara ve aylara g√∂re film daƒüƒ±lƒ±mlarƒ±
movie_counts.pivot_table(values = "count", index = "year_added", columns = "month_added", fill_value = 0)
```

```python
heatmap_film = movie_counts.pivot_table(values = "count", index = "year_added", columns = "month_added", fill_value = 0)
plt.figure(figsize = (16,10))
sns.heatmap(heatmap_film, annot = True, fmt = "d", cmap = "coolwarm")
plt.title("Yƒ±llara ve Aylara G√∂re Film Daƒüƒ±lƒ±mƒ±", fontsize = 20)
plt.show()
```

```python
# yƒ±llara g√∂re type durum
df.groupby(["release_year", "type"]).size().reset_index(name="count")
```

```python
release_counts = df.groupby(["release_year", "type"]).size().reset_index(name="count")

sns.lineplot(data=release_counts, x="release_year",
             y="count", hue="type")
plt.title("Her Yƒ±l √áƒ±kan Dizi & Film Sayƒ±sƒ±", fontsize = 16)
plt.show()
```

```python
content_counts = df.groupby(["year_added", "type"]).size().reset_index(name = "count")

sns.barplot(x = "year_added", y = "count", data = content_counts, hue = "type")
plt.xticks(rotation = 45)
plt.title("Netflix'e Her Yƒ±l Eklenen Dizi & Film Sayƒ±sƒ±",fontsize = 14)
plt.show()
```

```python
# 2019 yƒ±lƒ± Sci-Fi & Fantasy t√ºr√º yapƒ±mlar 
df[(df["release_year"] == 2019) & (df["listed_in"].str.contains("Sci-Fi & Fantasy"))][:5]
```

```python
Sci_Fi_Fantasy = df[(df["release_year"] == 2019) & (df["listed_in"].str.contains("Sci-Fi & Fantasy"))]

plt.figure(figsize = (12,6))
Sci_Fi_Fantasy["type"].value_counts().plot(kind = "pie", autopct = "%1.2f")
plt.title("Sci-Fi & Fantasy T√ºr√ºn√ºn Film ve Show Daƒüƒ±lƒ±mƒ±(2019)",fontsize = 18)
plt.show()
```
